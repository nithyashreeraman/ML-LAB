```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('iris.csv')

attr=df.columns.values[0:-1]
target=df[df.columns.values[-1]].values.tolist()
data=df[attr]
```

```python
data=np.array(data)
```

```python
# define distance function
def distance_function(x, y):
    return np.sqrt(np.sum((x - y) ** 2))

# compute pairwise distance matrix
D = squareform(pdist(data, distance_function))

# perform agglomerative clustering using single linkage
Z = linkage(D, method='single')
```

```python
# plot dendrogram
fig, ax = plt.subplots(figsize=(10, 6))
dendrogram(Z, labels=target)
ax.set_title('Dendrogram')
ax.set_xlabel('Samples')
ax.set_ylabel('Distance')
plt.show()
```
